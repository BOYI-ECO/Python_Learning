```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
a= np.linspace(0,10,1000)
b = np.sin(a)
c = np.cos(a**2)
plt.figure(figsize=(8,6))
plt.plot(a,b,"r-",label="sin(x)")
plt.plot(a,c,"g--",label="$cos(x^2)$")  #两个美元中的内容用数学表达
plt.legend(loc=2)
plt.xlabel("rainfall")
plt.ylabel("temp")
plt.title("exp1")
plt.savefig("function.jpg")

```
