```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
a= np.linspace(0,10,1000)
b = np.sin(a)
c = np.cos(a**2)
plt.figure(figsize=(8,6))
plt.plot(a,b,"r-",label="sin(x)")
plt.plot(a,c,"g--",label="$cos(x^2)$")  #
plt.legend(loc=2)
plt.xlabel("rainfall")
plt.ylabel("temp")
plt.title("exp1")
plt.savefig("function.jpg")

```

```python
y,x=np.ogrid[-2:2:200j,-2:2:200j]               ###  200j  ogrid/meshgrid/mgrid 
z= x*np.exp(-x**2-y**2)
plt.contour(z,10,color="b")
plt.imshow(z)
plt.colorbar()
```

```python
import mpl_toolkits.mplot3d
y,x=np.ogrid[-2:2:20j,-2:2:20j]
z= x*np.exp(-x**2-y**2)
ax = plt.subplot(111,projection='3d')
ax.plot_surface(x,y,z,rstride=2,cstride=3,color='r')
```
